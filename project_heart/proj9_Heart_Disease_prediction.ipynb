```python
x = heart_data.drop(columns=['target'])
y = heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

### Splitting Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, shuffle = True, stratify = y, random_state= 44)
```

```python
print(x_train.shape, x_test.shape , y_train.shape , y_test.shape)
```

### Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

### Model Evaluation

```python
y_train_predict = model.predict(x_train)
accuracy_of_trainig_data = accuracy_score(y_train, y_train_predict)
print("Accuracy of training data: ", accuracy_of_trainig_data)
```

```python
y_test_predict = model.predict(x_test)
accuracy_of_test_data = accuracy_score(y_test, y_test_predict)
print("Accuracy of test data: ", accuracy_of_test_data)
```

### Making a Predictive system

```python
input_data = (48,1,0,130,256,1,0,150,1,0,2,2,3)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
predicton = model.predict(input_data_reshaped)
print(predicton)

if (predicton[0]== 0):
    print("This person doesn't have a heart disease")
else:
    print("This person have a heart disease..!")
```
